### Geospatial Data

Geospatial data is map-based (although what constitutes a "map" can go beyond the familiar). Ultimately, geographic data  occurs either in bitmap (raster) form or vector (resizeable polygons). It is not unusual for the two kinds of data to be used together.

Because borders are often complex, following river courses, mountain ranges, and so on, the files that describe their geometries can be large. To reduce file sizes and image rendering times, we prefer to work with polygons that are no more detailed than the maximum resolution we need for our task. For this reason, geospatial files are often available at several different resolutions.

There are several popular options for working with maps in Python. We will use folium for its breadth and its use of open-source data.

```python
import folium
```

Using folium we can easily create a map. As it's SF data, by default we are using SF location.

```python
def generateBaseMap(default_location=[37.76, -122.45], default_zoom_start=12):
    '''
    Create a base map
    '''
    base_map = folium.Map(
        location = default_location
        , control_scale = True
        , zoom_start = default_zoom_start
    )

    return base_map
```

```python
base_map = generateBaseMap(default_location=[37.76, -122.45])
display(base_map)
```

Add a marker in the map using `Folium.Marker` for first 100 rows of the dataset.

```python
# ANSWER
# Check the columns to find latitude and longitude columns
print(df.columns)

# Assuming that the columns are 'Y' for latitude and 'X' for longitude (adjust as necessary)
latitude_column = 'Y'  # Replace with actual column name for latitude
longitude_column = 'X'  # Replace with actual column name for longitude

# Generate base map centered around San Francisco
sf_map = generateBaseMap()

# Add markers for the first 100 rows of the dataset
for i in range(100):
    lat = df[latitude_column].iloc[i]  # Latitude
    lon = df[longitude_column].iloc[i]  # Longitude
    
    # Create a popup with the incident number
    popup_info = f"Incident {i+1}: {df.iloc[i]['Category']}"  # Adjust to include any other information you want in the popup
    
    # Add a marker for each incident in the first 100 rows
    folium.Marker(
        location=[lat, lon],  # Position of the marker
        popup=popup_info  # Information displayed when clicking on the marker
    ).add_to(sf_map)

# Save the map as an HTML file
sf_map.save("sf_crimes_map_with_markers.html")

# Display the map 
sf_map
```

Add a circle in the map using `Folium Circle` for first 100 rows of the dataset.

```python
# ANSWER
# Assuming that the columns are 'Y' for latitude and 'X' for longitude (adjust as necessary)
latitude_column = 'Y'  # Replace with actual column name for latitude
longitude_column = 'X'  # Replace with actual column name for longitude

# Generate base map centered around San Francisco
sf_map = generateBaseMap()

# Add circles for the first 100 rows of the dataset
for i in range(100):
    lat = df[latitude_column].iloc[i]  # Latitude
    lon = df[longitude_column].iloc[i]  # Longitude
    
    # Create a popup with the incident number
    popup_info = f"Incident {i+1}: {df.iloc[i]['Category']}"  # Adjust to include any other information you want in the popup
    
    # Add a circle for each incident in the first 100 rows
    folium.Circle(
        location=[lat, lon],  # Position of the circle
        radius=200,  # Circle radius in meters
        color='red',  # Border color
        fill=True,  # Fill the circle
        fill_color='blue',  # Fill color
        fill_opacity=0.4,  # Fill opacity
        popup=popup_info  # Information displayed when clicking on the circle
    ).add_to(sf_map)

# Save the map as an HTML file
sf_map.save("sf_crimes_map_with_circles.html")

# Display the map 
sf_map
```
